```python
#Handling Outliers - Method2
def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x

auto_n=auto_n.apply(lambda x: outlier_capping(x))
```

```python
#Handling missings - Method2
def Missing_imputation(x):
    x = x.fillna(x.mean())
    return x

auto_n=auto_n.apply(lambda x: Missing_imputation(x))
```

```python
pandas_profiling.ProfileReport(auto_n)
```

```python
# An utility function to create dummy variable
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df
```

```python
#for c_feature in categorical_features
#House_Price_cat=HouseP[['Brick', 'Neighborhood']]

for c_feature in auto_cat.columns:
    auto_cat[c_feature] = auto_cat[c_feature].astype('category')
    auto_cat = create_dummies(auto_cat , c_feature )
```

```python
auto_new = pd.concat([auto_n, auto_cat], axis=1)
```
